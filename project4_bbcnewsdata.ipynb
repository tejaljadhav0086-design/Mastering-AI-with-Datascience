```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/bbc_news_data - bbc_news_data (1).csv')   #load dataset
```

```python
# number of columns
print("Number of columns:", df.shape[1])
```

```python
#checking missing values
print(df.isnull().sum)
```

```python
#checking datatypes
print(df.dtypes)
```

```python
#calculating mean numeric columns
print(df.mean(numeric_only=True))
```

```python
#correlation between columns
print(df.corr(numeric_only=True))
```

```python
#grouped data by categorical columns
grouped_data = df.groupby('Summary').mean(numeric_only=True)
print(grouped_data)
```

```python
#grouped data by categorical columns
grouped_data = df.groupby('Text').mean(numeric_only=True)
print(grouped_data)
```

```python
#performing EDA
df.describe()
```

```python
#correlation
plt.figure(figsize=(6, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
df.hist(figsize=(6, 8), bins=25)
plt.tight_layout()
plt.show()
```

```python
#Boxplot
df.boxplot(figsize=(6, 8))
plt.show()
```
